```python
import torch
import onnx
from ultralytics import YOLO
from ibm_watson_machine_learning import APIClient

# 1. Load the YOLOv8 model
model = YOLO('/content/drive/My Drive/models/Best_1.pt')  # Path to your trained YOLOv8 model

# 2. Export the model to ONNX format
onnx_model_path = '/content/drive/My Drive/models/Best_1.onnx'
try:
    model.export(format='onnx', imgsz=[640, 640], opset=12, dynamic=True, simplify=True)
    print(f"Model successfully exported to ONNX format at {onnx_model_path}")
except Exception as e:
    print(f"Failed to export model to ONNX: {e}")

# 3. Validate the ONNX model
try:
    onnx_model = onnx.load(onnx_model_path)
    onnx.checker.check_model(onnx_model)
    print("The ONNX model is valid.")
except Exception as e:
    print(f"ONNX model validation failed: {e}")

# 4. Initialize the IBM Watson Machine Learning client
wml_credentials = {
    'url': 'https://eu-de.ml.cloud.ibm.com',  # Replace with your IBM Cloud region URL if needed
    'apikey': 'iWL53mgvoUp5zE_z-VhBGKBESqwiZ3Ox0vFDe2RYYSz8'  # Replace with your API key
}
wml_client = APIClient(wml_credentials)

# 5. Set the default space
try:
    wml_client.spaces.list()  # List all spaces to find the GUID
    space_uid = '1f5c6bd4-dcdc-4d25-aab8-8b790e9e0034'  # Replace <SPACE_GUID> with your specific space GUID
    wml_client.set.default_space(space_uid)
    print(f"Default space set to: {space_uid}")
except Exception as e:
    print(f"Failed to set default space: {e}")

# 6. Define model properties for WML
try:
    software_spec_uid = wml_client.software_specifications.get_uid_by_name("default_py3.7")  # Ensure compatibility
    model_props = {
        wml_client.repository.ModelMetaNames.NAME: "Best_1",  # Name of the model
        wml_client.repository.ModelMetaNames.TYPE: "onnx_1.2",  # Correct type for ONNX models
        wml_client.repository.ModelMetaNames.SOFTWARE_SPEC_UID: software_spec_uid
    }
except Exception as e:
    print(f"Error fetching software specification UID: {e}")

# 7. Store the ONNX model in Watson Machine Learning
try:
    model_details = wml_client.repository.store_model(
        model=onnx_model_path,  # Path to the ONNX model
        meta_props=model_props
    )
    print("Model stored successfully:", model_details)
except Exception as e:
    print(f"Failed to store the model in WML: {e}")
```
